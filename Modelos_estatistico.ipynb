```python
# Excluíndo multiplas colunas
df = df.drop(['codigo_localidade', 'codigo_bairro'], axis=1)
df.head()
```

```python
# Vamos analisar a relação entre a variável de entrada are_m2 e a variável alvo valor_aluguel
sns.scatterplot(data = df, x = 'area_m2', y = 'valor_aluguel')
```

```python
df.head()
```

```python
# Definir a variável dependente (variável alvo)
y = df['valor_aluguel']
```

```python
# Definir a variável independente (variável preditora)
X = df['area_m2']
```

```python
# O Statsmodels requer a adição de uma constante à variável independente
X = sm.add_constant(X)
```

```python
# Criando o modelo
modelo = sm.OLS(y, X)
```

```python
# Treinamento do modelo
resultado = modelo.fit()
```

```python
print(resultado.summary())
```

```python
plt.figure(figsize = (12, 8))
plt.xlabel('area_m2', size = 16)
plt.ylabel('valor_aluguel', size = 16)
plt.plot(X['area_m2'], y, 'o', label = 'Dados Reais')
plt.plot(X['area_m2'], resultado.fittedvalues, 'r-', label = "LInha de Regressão (Previsões do Modelo)")
plt.legend(loc = 'best')
plt.show()
```
