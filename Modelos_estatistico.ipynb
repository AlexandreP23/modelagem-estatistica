```python
# Exclu√≠ndo multiplas colunas
df = df.drop(['codigo_localidade', 'codigo_bairro'], axis=1)
df.head()
```

```python
# Vamos analisar a rela√ß√£o entre a vari√°vel de entrada are_m2 e a vari√°vel alvo valor_aluguel
sns.scatterplot(data = df, x = 'area_m2', y = 'valor_aluguel')
```

```python
df.head()
```

```python
# Definir a vari√°vel dependente (vari√°vel alvo)
y = df['valor_aluguel']
```

```python
# Definir a vari√°vel independente (vari√°vel preditora)
X = df['area_m2']
```

```python
# O Statsmodels requer a adi√ß√£o de uma constante √† vari√°vel independente
X = sm.add_constant(X)
```

```python
# Criando o modelo
modelo = sm.OLS(y, X)
```

```python
# Treinamento do modelo
resultado = modelo.fit()
```

```python
print(resultado.summary())
```

```python
plt.figure(figsize = (12, 8))
plt.xlabel('area_m2', size = 16)
plt.ylabel('valor_aluguel', size = 16)
plt.plot(X['area_m2'], y, 'o', label = 'Dados Reais')
plt.plot(X['area_m2'], resultado.fittedvalues, 'r-', label = "LInha de Regress√£o (Previs√µes do Modelo)")
plt.legend(loc = 'best')
plt.show()
```
